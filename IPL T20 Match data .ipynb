```python
from selenium.common.exceptions import NoSuchWindowException,NoSuchElementException
Session=[]
Match_Number=[]
Venue=[]
Date_and_Time=[]
win=[]
First_Team=[]
Second_Team=[]
Link=[]

for i in range(2008,2024):
    url = f'https://www.iplt20.com/matches/results/{i}'
    driver=webdriver.Chrome()

    driver.get(url)
    # time.sleep(10)
    # session=driver.find_elements("xpath","//div[@class='customSelecBox openSBox']/div/div")
    match_no=driver.find_elements("xpath","//span[@class='vn-matchOrder ng-binding ng-scope']")
    venue=driver.find_elements("xpath","//div[@class='w50 fl']/span/p")
    date_time=driver.find_elements("xpath","//div[@class='vn-matchDateTime ng-binding']")
    win_t=driver.find_elements("xpath","//div[@class='vn-ticketTitle ng-binding ng-scope']")
    first_Team=driver.find_elements("xpath","//div[@class='vn-shedTeam']/div")
    second_team=driver.find_elements("xpath","//div[@class='vn-shedTeam vn-team-2']/div")
    lin=driver.find_elements("xpath","//div[@class='vn-ticnbtn']/a") 

    for i in range(len(match_no)):
        mat=match_no[i].text
        Match_Number.append(mat)
        ven=venue[i].text
        Venue.append(ven)
        date_t=date_time[i].text
        Date_and_Time.append(date_t)
        win_=win_t[i].text
        win.append(win_)
        first=first_Team[i].text
        First_Team.append(first)
        second=second_team[i].text
        Second_Team.append(second)
        li=lin[i].get_attribute("href")
        Link.append(li)


data={"Match Number":Match_Number,
      "Venue_":Venue,
      "Date and Time":Date_and_Time,
      "Win Team":win,
      "First Team":First_Team,
      "Second Team":Second_Team,
      "Link Match":Link


      }

df=pd.DataFrame(data)
print(df.to_string())
df.to_csv("ipl data.csv",index=False)
driver.quit()
```
